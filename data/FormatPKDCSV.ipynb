```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('ceidg_data_formated.csv', encoding='utf8')

data.head()
```

```python
data = data[data['Terminated'] == 1].drop(columns=['Terminated', 'DurationOfExistenceInMonths'] )

data.head()
```

```python
data['MonthOfTermination'] = pd.to_datetime(data['DateOfTermination'], format='%Y-%m-%d').dt.month

data.head()
```

```python
data['DateOfTermination'] = (data['YearOfTermination'] + data['MonthOfTermination'] / 12).round(2)

data.head()
```

```python
data = data.drop(columns=['MonthOfTermination', 'YearOfTermination'])

data.head()
```

```python
cols = data.columns[:-1].tolist()

cols
```

```python
data_grouped = data.groupby(cols).sum().reset_index()

data_grouped.head()
```

```python
print('before:', len(data))
print('after:', len(data_grouped))
```

```python
 data_crosstab = pd.crosstab(
    [data_grouped['MainAddressVoivodeship'], data_grouped['PKDMainSection'], data_grouped['PKDMainDivision']], 
    data_grouped['DateOfTermination'],
    values=data_grouped['Count'],
    aggfunc=sum
).fillna(0).reset_index()
    
del data_crosstab.columns.name
data_crosstab
```

```python
data_crosstab.to_csv('terminated_timeline.csv', index=False, encoding='utf8')
```
