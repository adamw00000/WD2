```python
import pandas as pd
import numpy as np
```

```python
pkd = pd.read_csv("pkd.csv", sep=";", encoding='utf8')

pkd.head()
```

```python
pkd = pkd[['symbol', 'nazwa']]

pkd.head()
```

```python
pkd['typ'] = \
    np.where(pkd.symbol.str.match(r'^[A-Z]+$'),
             'PKDMainSection', \
             np.where(pkd.symbol.str.match(r'^[0-9]+$'),
                      'PKDMainDivision', \
                      np.where(pkd.symbol.str.match(r'^[0-9]+.[0-9]$'),
                               'PKDMainGroup', \
                               'PKDMainClass')
                      ))

pkd.head()
```

```python
pkd = pkd[pkd['typ'].isin(['PKDMainSection', 'PKDMainDivision'])]

pkd.head()
```

```python
pkd['symbol'] = pkd.symbol.str.replace(r'^0+', '').str.replace(r'\.', '')

pkd.head()
```

```python
pkd.to_csv("pkd_data.csv", index=False, encoding='utf8')
```
