```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("ceidg_data_surv.csv", encoding="utf-8")
raw_data = data
data.head()
```

```python
data.columns.tolist()
```

```python
data = data[[
    'DateOfTermination',
    'Terminated',
    'DurationOfExistenceInMonths',
    'MainAddressVoivodeship',
    'PKDMainSection',
    'PKDMainDivision',
]]

data.head()
```

```python
for v in np.sort(data['MainAddressVoivodeship'].astype(str).unique()): print(v)
```

```python
data['MainAddressVoivodeship'] = data['MainAddressVoivodeship'].str\
    .lower()\
    .replace('ą', 'a', regex=True)\
    .replace('ć', 'c', regex=True)\
    .replace('ę', 'e', regex=True)\
    .replace('ł', 'l', regex=True)\
    .replace('ń', 'n', regex=True)\
    .replace('ó', 'o', regex=True)\
    .replace('ś', 's', regex=True)\
    .replace('ź', 'z', regex=True)\
    .replace('ż', 'z', regex=True)

for v in np.sort(data['MainAddressVoivodeship'].astype(str).unique()): print(v)
```

```python
data['MainAddressVoivodeship'] = data['MainAddressVoivodeship']\
.replace(['kuj pomorskie', 'kujawskopomorskie'], 'kujawsko-pomorskie')\
.replace(['maloplskie', 'malopolska'], 'malopolskie')\
.replace(['mazopwieckie'], 'mazowieckie')\
.replace(['pomorski'], 'pomorskie')\
.replace(['swietoktrzyskie'], 'swietokrzyskie')\
.replace(['warminsko-mazurskiego', 'warmi�sko-mazurskie', 'warm-maz'], 'warminsko-mazurskie')\
.replace(['wielkopolska', 'wielkopolski', 'wielskopolskie'], 'wielkopolskie')\
.replace(['zachodniopomorski', 'zachodno-pomorskie', 'zachondiopomorskie'], 'zachondiopomorskie')

for v in np.sort(data['MainAddressVoivodeship'].astype(str).unique()): print(v)
```

```python
voivodes = ['dolnośląskie',
 'kujawsko-pomorskie',
 'lubelskie',
 'lubuskie',
 'łódzkie',
 'małopolskie',
 'mazowieckie',
 'opolskie',
 'podkarpackie',
 'podlaskie',
 'pomorskie',
 'śląskie',
 'świętokrzyskie',
 'warmińsko-mazurskie',
 'wielkopolskie',
 'zachodniopomorskie']

vdict = {
    v.replace('ą', 'a')\
    .replace('ć', 'c')\
    .replace('ę', 'e')\
    .replace('ł', 'l')\
    .replace('ń', 'n')\
    .replace('ó', 'o')\
    .replace('ś', 's')\
    .replace('ź', 'z')\
    .replace('ż', 'z') : v
    for v in voivodes
}

vdict
```

```python
data['MainAddressVoivodeship'] = data['MainAddressVoivodeship'].replace(vdict)

for v in np.sort(data['MainAddressVoivodeship'].astype(str).unique()): print(v)
```

```python
data['MainAddressVoivodeship'] = data[data['MainAddressVoivodeship'].isin(voivodes)]['MainAddressVoivodeship']

for v in np.sort(data['MainAddressVoivodeship'].astype(str).unique()): print(v)
```

```python
data.head()
```

```python
data['DateOfTermination'] = data['DateOfTermination'] .fillna('2021-01-01')

data.head()
```

```python
data['DateOfTermination'] = pd.to_datetime(data['DateOfTermination'], format='%Y-%m-%d')

data['YearOfTermination'] = data['DateOfTermination'].dt.year

data.head()
```

```python
data = data.groupby(list(data.columns)).size().reset_index()

data.head()
```

```python
data = data.rename(columns={0: 'Count'})

data.head()
```

```python
print('before:', len(raw_data))
print('after:', len(data))
```

```python
data.to_csv('ceidg_data_formated.csv', index=False, encoding="utf-8")
```

```python
voivode_data = data[[
    'MainAddressVoivodeship',
    'Terminated',
    'YearOfTermination',
    'Count'
]]

voivode_data.head()
```

```python
voivode_data_terminated = voivode_data[voivode_data['Terminated'] == 1]\
    .drop(columns=['Terminated'])
voivode_data_terminated.head()
```

```python
voivoides_terminated_by_year = pd.crosstab(
    voivode_data_terminated['MainAddressVoivodeship'], 
    voivode_data_terminated['YearOfTermination'],
    values=voivode_data_terminated['Count'],
    aggfunc=sum
).reset_index()

del voivoides_terminated_by_year.columns.name
voivoides_terminated_by_year
```

```python
years = range(2011, 2021)

cumsum = voivoides_terminated_by_year[years].cumsum(axis=1)[years]

cumsum
```

```python
cumsum.insert(
    loc=0,
    column='MainAddressVoivodeship',
    value=voivoides_terminated_by_year['MainAddressVoivodeship']
)

voivoides_terminated_by_year = cumsum
voivoides_terminated_by_year
```

```python
all_voivoides = voivode_data[['MainAddressVoivodeship', 'Count']]\
    .groupby('MainAddressVoivodeship').sum()\
    .reset_index()\
    .rename(columns={'Count': 'all'})

all_voivoides
```

```python
voivoides_df = voivoides_terminated_by_year.merge(
    all_voivoides,
    on='MainAddressVoivodeship',
).drop(columns=['all_x']).rename(columns={'all_y': 'all'})

voivoides_df
```

```python
voivodes_lon_lat = pd.read_csv('voi_lon_lat.csv')

voivodes_lon_lat
```

```python
voivoides_df_final = voivoides_df.merge(
    voivodes_lon_lat,
    left_on='MainAddressVoivodeship',
    right_on='voivodeship'
).drop(columns=['voivodeship'])

voivoides_df_final
```

```python
voivoides_df_final.to_csv('voivoides_by_year.csv', index=False, encoding='utf8')
```
